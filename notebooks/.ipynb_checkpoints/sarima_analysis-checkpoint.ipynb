```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from numpy import loadtxt
from dateutil.relativedelta import relativedelta
import seaborn as sns
import sys
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf  
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.seasonal import seasonal_decompose
import sys
sys.path.append('..')
```

```python
df = pd.read_csv('../darwin.txt', index_col=0)
df.index.name='output'
df.reset_index(inplace=True)
```

```python
start = datetime.datetime.strptime("1900-01-01", "%Y-%m-%d")
date_list = [start + relativedelta(months=x) for x in range(0,1399)]
df['index'] =date_list
df.set_index(['index'], inplace=True)
df.index.name=None
df = df.head(120)
```

```python
df.plot(figsize=(12,8), title= 'Monthly Pressure', fontsize=14)
plt.savefig('month_pressure.png', bbox_inches='tight')
# plt.show();
```

```python
decomposition = seasonal_decompose(df.output, freq=12)  
fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(15, 8)
# plt.show();
```

```python
df.output_log= df.output.apply(lambda x: np.log(x))  
df['first_difference'] = df.output - df.output.shift(1)
df['seasonal_difference'] = df.output - df.output.shift(12)
df['seasonal_first_difference'] = df.first_difference - df.first_difference.shift(4)
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df.seasonal_first_difference.iloc[13:], lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df.seasonal_first_difference.iloc[13:], lags=40, ax=ax2)
plt.show();
```
