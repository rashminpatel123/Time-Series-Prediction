```python
from sklearn.svm import SVR
import numpy as np
from numpy import loadtxt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import sys
sys.path.append('..')
db = loadtxt('../darwin.txt', delimiter=',')
db = [round(i,1) for i in db]
```

```python
num_features = 5
X = list()
y = db[num_features:len(db)]
for i in range(len(db)-num_features):
    p = list()
    for j in range(num_features):
        p.append(db[i+num_features-j])
    X.append(p)
```

```python
X_train = X[:1299]
y_train = y[:1299]
X_test = X[1300:]
y_test = y[1300:]
```

```python
clf = SVR(C=1.0, epsilon=0.2)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
plt.plot(range(len(y_test)), y_test,'r');
plt.plot(range(len(y_pred)), y_pred,'b');
plt.show();
```
